```python
categorical_data = pd.get_dummies(df, drop_first = True, dtype = "int64")
categorical_data = categorical_data.rename(columns={"VehicleType_Car":"Car",
                                                   "VehicleType_LDV" :"LDV",
                                                   "VehicleType_Minibus": "Minibus",
                                                   "VehicleType_Motorcycle":"Motorcycle"
                                                ,"VehicleType_Other":"Other",
                                                   "VehicleType_Taxi":"Taxi",
                                                   "VehicleType_Toyota Carrolla":"Toyota Carrolla",
                                                   "VehicleType_Truck":"Truck",
                                                  "VehicleType_lf":"lf" })
categorical_data
```

```python
dy = categorical_data.describe()
categorical_data.mean()
categorical_data.mean().plot(x = 'Vehicle Type', y = 'Avergae Injuries', kind='bar', figsize=(8, 6))
plt.title('Injuries and Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Avergae Injuries')
```
